```python
from scipy.integrate.odepack import odeint
```

```python
# create a time array
t = np.linspace( 0.0, 10.0, 1000 )
```

```python
def force( t ):  
   amp = 1.0; freq = 1; phase = 0;
   return amp*sin( 2*pi*freq*t+phase )
```

```python
# solve the linear ode
linear_init = [1]  # -- inital condition

def linear_eq( y,t ): 
    return [-y[0]]

linear = odeint( linear_eq, linear_init, t)

# solve the linear ode
def linearf_eq( y,t ):
    return [ -y[0]+force(t) ]
linearf = odeint( linearf_eq, linear_init, t)
```

```python
# solve the non-linear ode
nonlinear_init = [1] # -- initial condition

def nonlinear_eq( y,t ): 
    return [-y[0]*y[0]]
nonlinear = odeint( nonlinear_eq, nonlinear_init, t)

# solve the non-linear ode
def nonlinearf_eq( y,t ):
    return [ -y[0]*y[0]+force(t) ]
nonlinearf = odeint( nonlinearf_eq, nonlinear_init, t)
```

```python
# plotting
p1, = plot( t, linear[:,0], color='blue' )
p2, = plot( t, nonlinear[:,0], color='red'  )
p3, = plot( t, linearf[:,0], color='blue',linestyle='--' )
p4, = plot( t, nonlinearf[:,0], color='red',linestyle='--'   )

xlabel('time (sec)')
ylabel('output')
title('Natural Responses')
legend([p1,p2], ['Linear', 'Non-linear'])
```
